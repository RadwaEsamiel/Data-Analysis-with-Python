```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cleaned_absenteeism_data.csv")
```

```python
visualization_df = data.copy()
```

```python
visualization_df.columns.values
```

```python
plt.bar(x= visualization_df['Education'],
       height =  visualization_df['Absenteeism Time in Hours']
       )
plt.show()
```

```python
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (9,6))

plt.bar(x= visualization_df['Day of Week'],
       height =  visualization_df['Absenteeism Time in Hours'],
        color = "midnightblue"
       )
plt.xticks(rotation = 45)
plt.title("Absenteeism Time by each Day of the Week.")
plt.show()
```
